# Dropping the Unnecessary Columns:
## all the id columns are unncessary because their names are existing in the data

```python
health_data= health_data.drop(columns=[
    'measure_id',
    'location_id',
    'sex_id',
    'age_id',
    'cause_id',
    'metric_id'
])
```

```python
health_data.head()
```

#Change Names of some Columns to match other data sets:

```python
health_data = health_data.rename(columns={
    'location_name': 'country',
    'measure_name': 'measure',
    'cause_name': 'disease',
    'val': 'value'
})
```

```python
health_data.columns
```

```python
health_data = health_data.rename(columns={
    'sex_name': 'sex',
    'age_name': 'age',
    'metric_name': 'metric'
})
```
