```python
import pandas as pd
AirQuality=pd.read_excel("who_ambient_air_quality_database_version_2024_(v6.1).xlsx", sheet_name="Update 2024 (V6.1)")
```

```python
AirQuality.isnull().sum()
```

```python
AirQuality['who_region'] = AirQuality['who_region'].astype(str)
```

```python
AirQuality['iso3'] = AirQuality['iso3'].astype(str)
```

```python
AirQuality['country_name'] = AirQuality['country_name'].astype(str)
```

```python
AirQuality['city'] = AirQuality['city'].astype(str)
```

```python
AirQuality[AirQuality['year'].isnull()]
```

```python
AirQuality = AirQuality.dropna(subset=['year'])
```

```python
AirQuality.shape
```

```python
AirQuality['year'] = AirQuality['year'].astype(int)
```

```python
AirQuality['version'] = AirQuality['version'].astype(str)
```

```python
cols = ['pm10_concentration', 'pm25_concentration', 'no2_concentration',
       'pm10_tempcov', 'pm25_tempcov', 'no2_tempcov']

AirQuality[cols] = AirQuality.groupby('city')[cols].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
AirQuality.isnull().sum()
```

```python
AirQuality[['pm10_concentration', 'pm25_concentration', 'no2_concentration',
       'pm10_tempcov', 'pm25_tempcov', 'no2_tempcov']] = \
AirQuality[['pm10_concentration', 'pm25_concentration', 'no2_concentration',
       'pm10_tempcov', 'pm25_tempcov', 'no2_tempcov']].fillna(0)
```

```python
AirQuality.isnull().sum()
```

```python
AirQuality.shape
```

```python
AirQuality['type_of_stations'] = AirQuality['type_of_stations'].fillna('N/A')
```

```python
AirQuality.isnull().sum()
```

```python
AirQuality['type_of_stations'] = AirQuality['type_of_stations'].astype(str)
```

```python
AirQuality = AirQuality.drop('reference', axis=1)
AirQuality = AirQuality.drop('web_link', axis=1)
```

```python
AirQuality.isnull().sum()
```

```python
cols = ['population']

AirQuality[cols] = AirQuality.groupby('city')[cols].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
AirQuality.isnull().sum()
```

```python
AirQuality[['population']] = \
AirQuality[['population']].fillna(0)
```

```python
AirQuality.isnull().sum()
```

```python
AirQuality['population_source'] = AirQuality['population_source'].fillna('N/A')
```

```python
AirQuality.isnull().sum()
```

```python
AirQuality = AirQuality.drop('who_ms', axis=1)
```

```python
AirQuality['number_of_stations'] = AirQuality['type_of_stations'].apply(
    lambda x: 0 if x == 'N/A' else x.count(',') + 1
)
```

```python
cols = list(AirQuality.columns)
cols.insert(13, cols.pop(cols.index('number_of_stations')))
AirQuality = AirQuality[cols]
```

```python
AirQuality.isnull().sum()
```

```python
AirQuality.to_excel("AirQuality_cleaned.xlsx", index=False)
```
